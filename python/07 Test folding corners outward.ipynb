```python
# Colormap for plots
cmap = plt.cm.viridis
norm = matplotlib.colors.BoundaryNorm(np.linspace(0,4,9),cmap.N)
```

# 3x2 target grid with 6x4 topographic data containing a diagonal wall

Set cell means of fine (6x4 mesh) to source topography

```python
ni,nj = 4,4
FG = ThinWalls.ThinWalls(lon=np.linspace(0,ni,ni+1), lat=np.linspace(0,nj,nj+1))
FG.set_cell_mean(
    np.array(
[
    [0,0,0,1],
    [0,0,2,0],
    [0,0,1,0],
    [1,2,0,0]
]))
FG.set_edge_mean(
    np.array(
[
    [0,0,0,1,1],
    [0,0,2,2,0],
    [0,4,1,4,0],
    [1,2,2,0,0]
]),
    np.array(
[
    [0,0,0,1],
    [0,0,2,1],
    [0,0,1,0],
    [1,4,3,0],
    [1,2,0,0]
])
)

plt.figure(figsize=(16,4))
ax = plt.subplot(131); plt.colorbar( FG.plot(ax, norm=norm, cmap=cmap, metric='min'), ax=ax); plt.title('Actual min');
ax = plt.subplot(132); plt.colorbar( FG.plot(ax, norm=norm, cmap=cmap), ax=ax); plt.title('Actual mean');
ax = plt.subplot(133); plt.colorbar( FG.plot(ax, norm=norm, cmap=cmap, metric='max'), ax=ax); plt.title('Actual max');
```

Initialize effective values by setting to actual values

```python
FG.init_effective_values()

plt.figure(figsize=(16,4))
ax = plt.subplot(131); plt.colorbar(
    FG.plot(ax, norm=norm, cmap=cmap, measure='effective', metric='min'), ax=ax); plt.title('Effective min');
ax = plt.subplot(132); plt.colorbar(
    FG.plot(ax, norm=norm, cmap=cmap, measure='effective'), ax=ax); plt.title('Effective mean');
ax = plt.subplot(133); plt.colorbar(
    FG.plot(ax, norm=norm, cmap=cmap, measure='effective', metric='max'), ax=ax); plt.title('Effective max');
```

Coarsen

```python
FG.push_corners()

plt.figure(figsize=(16,4))FG.push_corners()

ax = plt.subplot(131); plt.colorbar(
    FG.plot(ax, norm=norm, cmap=cmap, measure='effective', metric='min'), ax=ax); plt.title('Effective min');
ax = plt.subplot(132); plt.colorbar(
    FG.plot(ax, norm=norm, cmap=cmap, measure='effective'), ax=ax); plt.title('Effective mean');
ax = plt.subplot(133); plt.colorbar(
    FG.plot(ax, norm=norm, cmap=cmap, measure='effective', metric='max'), ax=ax); plt.title('Effective max');
```
